```python
df['date'] = pd.to_datetime(df['date'], format = '%d/%m/%Y')
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month_name()
df['Day'] = df['date'].dt.day_name()
```

```python
df['Hour'] = pd.to_datetime(df['time'], format = '%H:%M', errors = 'coerce').dt.hour
df[['time', 'Hour']].head()
```

```python
# Analysis 1: Accident By Hour

plt.figure(figsize = (12,6))

sns.countplot(x='Hour', data=df, order=sorted(df['Hour'].dropna().unique()))
plt.title('Accident by Hour of the day')
plt.xlabel('Hour')
plt.ylabel('No. of Accidents')
plt.savefig('Accident_by_Hour_of_the_day.png')
plt.show()
print('Most accidents occur during morning and evening rush hours.')
```

```python
# Analysis 2: Accident By Day

plt.figure(figsize = (10,5))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

sns.countplot(x = 'Day', data = df, order = order)
plt.title("Accident By Day of the week")
plt.xticks(rotation = 45)
plt.savefig('Accident_By_Day_of_the_week.png')
plt.show()
print('Weekdays generally have more traffic-related accidents.')
```

```python
# Analysis 3: Collision Severity

plt.figure(figsize = (8,5))

sns.countplot(x = 'collision_severity', data = df)
plt.title('collision severity distribuition')
plt.savefig('collision_severity.png')
plt.show()
print('Slight collisions are usually more common than serious or fatal collisions.')
```

```python
# Analysis 4: Weather Conditions
top_weather = (df['weather_conditions'].value_counts().head(10))
plt.figure(figsize = (12,6))

top_weather.plot(kind='bar')

plt.title('Top Weather Conditios During Accidents')
plt.ylabel('Count')
plt.savefig('Top_Weather_Condition_During_Accidents.png')
plt.show()
print('Many accidents occur during fine weather due to increased traffic volume.')
```

```python
# Analysis 5: Road Surface Conditions
plt.figure(figsize=(10, 6))
road_counts = df['road_surface_conditions'].value_counts()

road_counts.plot(kind='bar')

plt.title("Road Surface Conditions")
plt.xlabel("Road Surface Condition")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('Road_Surface_Conditions.png')
plt.show()

```

```python
# Analysis 6: Urban Vs Rural

plt.figure(figsize = (6,5))
sns.countplot(x = 'urban_or_rural_area', data = df)
plt.title('Urban Vs Rural Area')
plt.show()
plt.savefig('urban_or_rural_area.png')
print('Urban areas generally experience more accidents because of higher traffic density.')
```

```python
# Analysis 7: Speed Limit Analysis
plt.figure(figsize = (10,5))
sns.countplot(x='speed_limit', data = df)

plt.title('Accidents By Speed Limit')
plt.xticks(rotation = 45)
plt.savefig('Accidents_By_Speed_Limit.png')
plt.show()
print('Accidents frequently occur on roads with common speed limits such as 30 mph.')
```

```python
# Analysis 8: Light Conditions
plt.figure(figsize = (12,5))
df['light_conditions'].value_counts().plot(kind = 'bar')
plt.title('Accidents By Light Conditions')
plt.ylabel('Count')
plt.savefig('Accidents_By_Light_Conditions.png')
plt.show()
print('Daylight conditions often show more accidents due to increased traffic.')
```

```python
# Analysis 9: Accident Hotspots
sample = df.sample(min(5000, len(df)), random_state = 42)

plt.figure(figsize = (10,8))
plt.scatter(sample['longitude'], sample['latitude'], alpha = 0.5)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig('Accident_Hotspots.png')
plt.show()
print('Clusters indicate accident-prone regions.')
```

```python
# Correlation Heatmap
numeric_cols = ['number_of_casualties', 'number_of_vehicles', 'speed_limit']

corr = df[numeric_cols].corr()

plt.figure(figsize = (6,4))

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('heatmap.png')
plt.show()
```
